````python
products = ["t-shirt", "mug", "book", "hat", "keychain"]

def initialize_inventory(products):
    inventory = {product: int(input(f"Enter quantity of {products} available: ")) for product in products}
    return inventory
inventory = initialize_inventory(products)
````

````python
def get_customer_orders():
    quantity_order = int(input("Enter the number of customer orders: "))
    customer_order = {input("Enter the product that a customer wants to order: ") for order in range(quantity_order) }
    return quantity_order, customer_order

quantity_order, customer_order = get_customer_orders()
````

````python
def products_price (customer_order):
    prices = [int(input("Please introduce the price of each product in the customer order: ")) for price in customer_order]
    total_price = float(sum(prices)) 
    percentage = len(customer_order) / len(products) * 100
    return prices, total_price, percentage

prices, total_price, percentage = products_price (customer_order)
````

````python
def update_inventory(customer_orders, inventory):
    updated_inventory = {product: (inventory[product] - 1 if product in customer_orders else inventory[product]) for product in inventory}
    del_product = {product: quantity for product ,quantity in updated_inventory.items() if quantity < 0} #product porque voy a iterar. Luego product,quantity porque es un diccionario con keys y values. Recuerdo poner item y ya luego el filtro de si es menor a cero.
    
    return updated_inventory, del_product

updated_inventory, del_product = update_inventory(customer_order, inventory)
                      
````
